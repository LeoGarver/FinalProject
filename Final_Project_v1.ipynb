## Use Merge to join the Pitcher data with the data from the team and season

```python
master= pd.merge(pitchers, bullpen, how='inner', on=['Season', 'Team'], suffixes=('', '_bullpen'))
```

```python
master.head()
```

```python
batting= pd.read_csv('data\Final_Project_Data\Team_Batting.csv',low_memory=False, usecols=['Season', 'Team', 'wOBA'])
```

```python
master= pd.merge(master, batting, how='inner', on=['Season', 'Team'], suffixes=('', '_team'))
```

```python
master.head()
```

## Create new fields to use in our analysis

```python
master['IP/GS'] = master['IP']/master['GS']
master['W/GS'] = master['W']/master['GS']
master['L/GS'] = master['L']/master['GS']
master['RA9'] = master['R']/master['IP']*9
master['BullpenRA9'] = master['R_bullpen']/master['IP_bullpen']*9
master['StarterR'] = master['RA9']*master['IP/GS']/9
master['BullpenR'] =  master['BullpenRA9']*(9-master['IP/GS'])/9
master['RAA'] = master['StarterR'] + master['BullpenR']
```

## Model #1 Linear Regression OLS with 4 features

```python
features_array = master.get(['RA9', 'IP/GS', 'wOBA', 'BullpenRA9'])
features_values = features_array.values
target = master['W/GS'].values
data = master.get(['RA9', 'IP/GS', 'wOBA', 'BullpenRA9', 'W/GS'])
```

```python
features_train, features_test, target_train, target_test = train_test_split(features_values, target, test_size=0.20, random_state=0)
```

```python
regr = linear_model.LinearRegression()
regr.fit(features_train, target_train)
```

```python
regr.score(features_test, target_test)
```

```python
regr.coef_
```

```python
from pandas.tools.plotting import scatter_matrix
scat = scatter_matrix(data, figsize = (5,5))
```

## Model #2 SGD Linear Regression with same features

```python
SGD = linear_model.SGDRegressor(loss='huber')
```

```python
SGD.fit(features_train, target_train)
SGD.score(features_test, target_test)
```

```python
losstarget = master['L/GS'].values
```

```python
features_train, features_test, target_train, target_test = train_test_split(features_values, losstarget, test_size=0.20, random_state=0)
```

```python
lossreg = linear_model.LinearRegression()
lossreg.fit(features_train, target_train)
```

```python
lossreg.coef_
```

```python
lossreg.score(features_test, target_test)
```

## Create new stats based on Model #1

```python
master['Pred W/GS'] = regr.predict(features_array)
master['Pred L/GS'] = lossreg.predict(features_array)
master['Pred W'] = master['GS']*master['Pred W/GS']
master['Pred L'] = master['GS']*master['Pred L/GS']
master['Luck'] = master['W']-master['Pred W']
```

```python
#master.sort(column = 'Pred W/GS', ascending=False)
print master.get(['Season', 'Team', 'Name', 'W', 'Pred W', 'Luck', 'L', 'Pred L', 'GS', 'Pred W/GS']).sort(column = 'Luck', ascending=True)
```
